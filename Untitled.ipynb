```python
import pandas as pd

# Load the dataset
file_path = 'dataset.xlsx'
data_ib = pd.read_excel(file_path, sheet_name='DATA IB')
data_ik = pd.read_excel(file_path, sheet_name='DATA IK')

# Display the first few rows of each sheet for analysis
data_ib_head = data_ib.head()
data_ik_head = data_ik.head()

data_ib_head, data_ik_head
```

```python
# Combine 'desc' from DATA IB and 'tujuan' from DATA IK into a single column "keperluan"
data_ib['keperluan'] = data_ib['desc'].str.strip()
data_ik['keperluan'] = data_ik['tujuan'].str.strip()

# Combine the two datasets
combined_data = pd.concat([data_ib[['keperluan']], data_ik[['keperluan']]], ignore_index=True)

# Normalize text: lowercasing and removing extra whitespaces
combined_data['keperluan'] = combined_data['keperluan'].str.lower().str.strip()

# Display the first few rows of the combined dataset
combined_data.head()
```

```python
from nltk.corpus import stopwords
import re

# Step 1: Remove null or empty values
combined_data.dropna(subset=['keperluan'], inplace=True)
combined_data = combined_data[combined_data['keperluan'].str.strip() != '']

# Step 2: Remove duplicates
combined_data.drop_duplicates(subset=['keperluan'], inplace=True)

# Step 3: Further normalization - remove special characters and numbers
combined_data['keperluan'] = combined_data['keperluan'].apply(
    lambda x: re.sub(r'[^a-z\s]', '', x)  # Retain only letters and spaces
)

# Step 4: Remove stopwords (optional, for text simplification)
stop_words = set(stopwords.words('indonesian'))  # Using NLTK's Indonesian stopwords
combined_data['keperluan'] = combined_data['keperluan'].apply(
    lambda x: ' '.join(word for word in x.split() if word not in stop_words)
)

# Display the preprocessed data
combined_data.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the preprocessed data (simulating the user's results)
combined_data = pd.DataFrame({
    'keperluan': [
        'bertemu orang tua',
        'libur',
        'libur semester',
        'memperbaiki laptop',
        'tugas proyek'
    ]
})
combined_data = pd.concat([combined_data] * (3439 // len(combined_data)), ignore_index=True)  # Simulate dataset size

# Step 1: Calculate frequency of each "keperluan"
frequency = combined_data['keperluan'].value_counts()

# Step 2: Visualize frequency with a bar chart
plt.figure(figsize=(10, 6))
frequency.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Keperluan Mahasiswa', fontsize=16)
plt.xlabel('Keperluan', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 3: Generate a word cloud for "keperluan"
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(combined_data['keperluan']))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Keperluan', fontsize=16)
plt.show()

# Summary of unique entries
unique_count = combined_data['keperluan'].nunique()
unique_count
```

```python
# Membuat fungsi untuk mengelompokkan keperluan
def group_keperluan(keperluan):
    if 'libur' in keperluan.lower() or 'semester' in keperluan.lower() or 'natal' in keperluan.lower() or 'paskah' in keperluan.lower() or 'lebaran' in keperluan.lower():
        return 'Libur'
    elif 'bertemu' in keperluan.lower() or 'keluarga' in keperluan.lower() or 'orang tua' in keperluan.lower() or 'rumah' in keperluan.lower() or 'pulang' in keperluan.lower():
        return 'Bertemu Keluarga'
    elif 'laptop' in keperluan.lower() or 'service' in keperluan.lower():
        return 'Memperbaiki Laptop'
    elif 'behel' in keperluan.lower() or 'gigi' in keperluan.lower() or 'kontrol' in keperluan.lower() or 'kesehatan' in keperluan.lower() or 'obat' in keperluan.lower():
        return 'Berobat'
    elif 'membeli' in keperluan.lower() or 'barang' in keperluan.lower() or 'pribadi' in keperluan.lower() or 'mengurus' in keperluan.lower() or 'keperluan' in keperluan.lower():
        return 'Urusan Pribadi'
    elif 'proyek akhir' in keperluan.lower() or 'requirement' in keperluan.lower() or 'data' in keperluan.lower():
        return 'Tugas Kuliah'
    else:
        return 'Lainnya'  # Biarkan keperluan lain tetap

# Terapkan fungsi ke kolom 'keperluan'
combined_data['keperluan_grouped'] = combined_data['keperluan'].apply(group_keperluan)

# Hitung frekuensi keperluan setelah pengelompokan
frequency_grouped = combined_data['keperluan_grouped'].value_counts()

# Visualisasi hasil pengelompokan dengan bar chart
plt.figure(figsize=(10, 6))
frequency_grouped.plot(kind='bar', color='skyblue')
plt.title('Frekuensi Keperluan Mahasiswa Setelah Pengelompokan', fontsize=16)
plt.xlabel('Keperluan (Dikelompokkan)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualisasi Word Cloud setelah pengelompokan
wordcloud_grouped = WordCloud(width=800, height=400, background_color='white').generate(' '.join(combined_data['keperluan_grouped']))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_grouped, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Grouped Keperluan', fontsize=16)
plt.show()
```
