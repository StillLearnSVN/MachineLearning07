```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data Processing and ML Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import PCA

# Clustering Models
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Additional Preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

class BusinessOpportunityAnalysis:
    def __init__(self, file_path):
        """
        Initialize the analysis with the dataset
        
        Parameters:
        file_path (str): Path to the Excel file
        """
        # 1. Data Loading and Preprocessing
        self.load_and_preprocess_data(file_path)
        
        # Download NLTK resources
        nltk.download('punkt', quiet=True)
        nltk.download('stopwords', quiet=True)
    
    def load_and_preprocess_data(self, file_path):
        """
        Load and preprocess data from Excel file
        """
        # Read Excel sheets
        ib_data = pd.read_excel(file_path, sheet_name='DATA IB')
        ik_data = pd.read_excel(file_path, sheet_name='DATA IK')
        
        # Combine purposes
        ib_data['source'] = 'IB'
        ik_data['source'] = 'IK'
        ib_data['purpose'] = ib_data['desc']
        ik_data['purpose'] = ik_data['tujuan']
        
        # Merge datasets
        self.combined_data = pd.concat([
            ib_data[['purpose', 'source']], 
            ik_data[['purpose', 'source']]
        ], ignore_index=True)
        
        # Remove NaN values
        self.combined_data = self.combined_data.dropna(subset=['purpose'])
    
    def advanced_text_preprocessing(self, text):
        """
        Advanced text preprocessing
        
        Parameters:
        text (str): Input text to preprocess
        
        Returns:
        str: Preprocessed text
        """
        # Convert to lowercase
        text = text.lower()
        
        # Remove special characters and digits
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenization
        tokens = word_tokenize(text)
        
        # Remove stopwords
        stop_words = set(stopwords.words('indonesian'))
        tokens = [token for token in tokens if token not in stop_words]
        
        return ' '.join(tokens)
    
    def feature_extraction(self, preprocessing_method='tfidf'):
        """
        Extract features from text data
        
        Parameters:
        preprocessing_method (str): Feature extraction method
        
        Returns:
        scipy.sparse matrix: Feature matrix
        """
        # Preprocess text
        self.combined_data['processed_purpose'] = self.combined_data['purpose'].apply(self.advanced_text_preprocessing)
        
        # Feature extraction
        if preprocessing_method == 'tfidf':
            vectorizer = TfidfVectorizer(stop_words='english')
        else:
            vectorizer = CountVectorizer(stop_words='english')
        
        # Fit and transform
        feature_matrix = vectorizer.fit_transform(self.combined_data['processed_purpose'])
        
        # Store feature names for interpretation
        self.feature_names = vectorizer.get_feature_names_out()
        
        return feature_matrix
    
    def perform_clustering(self, feature_matrix, n_clusters_range=range(2, 11), clustering_method='kmeans'):
        """
        Perform clustering with hyperparameter tuning
        
        Parameters:
        feature_matrix (scipy.sparse matrix): Input feature matrix
        n_clusters_range (range): Range of cluster numbers to evaluate
        clustering_method (str): Clustering algorithm to use
        
        Returns:
        dict: Clustering results and evaluation metrics
        """
        # Evaluation metrics storage
        evaluation_metrics = {
            'silhouette_scores': [],
            'calinski_scores': [],
            'davies_bouldin_scores': []
        }
        
        # Best clustering parameters
        best_params = {
            'n_clusters': None,
            'best_silhouette_score': -1
        }
        
        # Dimensionality reduction for visualization and performance
        pca = PCA(n_components=2)
        reduced_features = pca.fit_transform(feature_matrix.toarray())
        
        # Clustering methods
        for n_clusters in n_clusters_range:
            if clustering_method == 'kmeans':
                clusterer = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
            elif clustering_method == 'dbscan':
                clusterer = DBSCAN(eps=0.5, min_samples=3)
            
            # Fit clustering
            cluster_labels = clusterer.fit_predict(feature_matrix.toarray())
            
            # Evaluation metrics
            silhouette = silhouette_score(feature_matrix.toarray(), cluster_labels)
            calinski = calinski_harabasz_score(feature_matrix.toarray(), cluster_labels)
            davies = davies_bouldin_score(feature_matrix.toarray(), cluster_labels)
            
            # Store metrics
            evaluation_metrics['silhouette_scores'].append(silhouette)
            evaluation_metrics['calinski_scores'].append(calinski)
            evaluation_metrics['davies_bouldin_scores'].append(davies)
            
            # Update best parameters
            if silhouette > best_params['best_silhouette_score']:
                best_params['n_clusters'] = n_clusters
                best_params['best_silhouette_score'] = silhouette
            
            # Visualization
            plt.figure(figsize=(10, 6))
            plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=cluster_labels, cmap='viridis')
            plt.title(f'Clustering Visualization (n_clusters = {n_clusters})')
            plt.xlabel('PCA Component 1')
            plt.ylabel('PCA Component 2')
            plt.colorbar(label='Cluster')
            plt.tight_layout()
            plt.show()
        
        # Plot evaluation metrics
        plt.figure(figsize=(12, 4))
        plt.subplot(131)
        plt.plot(list(n_clusters_range), evaluation_metrics['silhouette_scores'])
        plt.title('Silhouette Score')
        plt.xlabel('Number of Clusters')
        
        plt.subplot(132)
        plt.plot(list(n_clusters_range), evaluation_metrics['calinski_scores'])
        plt.title('Calinski-Harabasz Score')
        plt.xlabel('Number of Clusters')
        
        plt.subplot(133)
        plt.plot(list(n_clusters_range), evaluation_metrics['davies_bouldin_scores'])
        plt.title('Davies-Bouldin Score')
        plt.xlabel('Number of Clusters')
        
        plt.tight_layout()
        plt.show()
        
        return {
            'best_params': best_params,
            'evaluation_metrics': evaluation_metrics
        }
    
    def generate_business_insights(self, feature_matrix, cluster_labels):
        """
        Generate business insights from clustered data
        
        Parameters:
        feature_matrix (scipy.sparse matrix): Input feature matrix
        cluster_labels (numpy.array): Cluster labels
        
        Returns:
        dict: Business opportunity insights
        """
        # Add cluster labels to dataframe
        self.combined_data['cluster'] = cluster_labels
        
        # Analyze clusters
        cluster_insights = {}
        for cluster in np.unique(cluster_labels):
            cluster_data = self.combined_data[self.combined_data['cluster'] == cluster]
            
            # Find top keywords for the cluster
            cluster_matrix = feature_matrix[cluster_labels == cluster]
            top_keyword_indices = cluster_matrix.sum(axis=0).argsort()[0, ::-1][:5]
            top_keywords = [self.feature_names[i] for i in top_keyword_indices.tolist()[0]]
            
            cluster_insights[cluster] = {
                'keywords': top_keywords,
                'sample_purposes': cluster_data['purpose'].sample(min(3, len(cluster_data))).tolist(),
                'business_opportunities': self.map_keywords_to_opportunities(top_keywords)
            }
        
        return cluster_insights
    
    def map_keywords_to_opportunities(self, keywords):
        """
        Map keywords to potential business opportunities
        
        Parameters:
        keywords (list): Top keywords for a cluster
        
        Returns:
        list: Potential business opportunities
        """
        opportunity_mappings = {
            'laptop': ['Computer Repair Shop', 'Laptop Sales and Service', 'Tech Accessory Store'],
            'orang': ['Family Support Services', 'Travel Assistance', 'Family Reunion Planning'],
            'market': ['Market Research Consultancy', 'Local Business Consulting', 'Research Services'],
            'ibadah': ['Event Management', 'Religious Event Planning', 'Community Event Services'],
            'dinas': ['Government Liaison Services', 'Permit and Documentation Assistance']
        }
        
        opportunities = []
        for keyword in keywords:
            if keyword in opportunity_mappings:
                opportunities.extend(opportunity_mappings[keyword])
        
        return list(set(opportunities))

def main():
    # File path
    file_path = 'dataset.xlsx'
    
    # Initialize analysis
    analysis = BusinessOpportunityAnalysis(file_path)
    
    # Feature Extraction
    feature_matrix = analysis.feature_extraction(preprocessing_method='tfidf')
    
    # Hyperparameter Tuning and Clustering
    clustering_results = analysis.perform_clustering(
        feature_matrix, 
        n_clusters_range=range(2, 11), 
        clustering_method='kmeans'
    )
    
    # Best number of clusters
    best_n_clusters = clustering_results['best_params']['n_clusters']
    
    # Perform final clustering with best parameters
    kmeans = KMeans(n_clusters=best_n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(feature_matrix.toarray())
    
    # Generate Business Insights
    business_insights = analysis.generate_business_insights(feature_matrix, cluster_labels)
    
    # Print Results
    print("\nBusiness Opportunity Analysis Results:")
    for cluster, insights in business_insights.items():
        print(f"\nCluster {cluster}:")
        print(f"Top Keywords: {insights['keywords']}")
        print("Sample Purposes:")
        for purpose in insights['sample_purposes']:
            print(f"- {purpose}")
        print("Potential Business Opportunities:")
        for opportunity in insights['business_opportunities']:
            print(f"- {opportunity}")

if __name__ == "__main__":
    main()
```
