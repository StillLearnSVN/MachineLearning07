```python
import pandas as pd
import matplotlib.pyplot as plt

class BusinessUnderstandingPhase:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df_ib = None
        self.df_ik = None
    
    def load_and_explore_data(self):
        """
        Load and explore the dataset
        """
        # Read Excel file
        self.df_ib = pd.read_excel(self.file_path, sheet_name='DATA IB')
        self.df_ik = pd.read_excel(self.file_path, sheet_name='DATA IK')
        
        # Basic information
        print("IB Data Information:")
        print(self.df_ib.info())
        print("\nIK Data Information:")
        print(self.df_ik.info())
        
        # Descriptive statistics
        print("\nIB Data Descriptive Statistics:")
        print(self.df_ib.describe())
        print("\nIK Data Descriptive Statistics:")
        print(self.df_ik.describe())
        
        # Visualization of top destinations/purposes
        plt.figure(figsize=(15, 6))
        
        plt.subplot(1, 2, 1)
        self.df_ib['desc'].value_counts().head(10).plot(kind='bar')
        plt.title('Top 10 IB Destinations')
        plt.xlabel('Destination')
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right')
        
        plt.subplot(1, 2, 2)
        self.df_ik['tujuan'].value_counts().head(10).plot(kind='bar')
        plt.title('Top 10 IK Destinations')
        plt.xlabel('Destination')
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right')
        
        plt.tight_layout()
        plt.show()
        
        # Save basic insights
        with open('business_understanding_insights.txt', 'w') as f:
            f.write("Business Opportunity Prediction Project\n")
            f.write("=====================================\n\n")
            f.write("Project Objective: Predict business opportunities\n")
            f.write("based on student permit purposes\n\n")
            f.write("Key Insights:\n")
            f.write(f"Total IB Entries: {len(self.df_ib)}\n")
            f.write(f"Total IK Entries: {len(self.df_ik)}\n")
    
    def save_processed_data(self):
        """
        Save processed data for next stages
        """
        if self.df_ib is not None and self.df_ik is not None:
            # Combine datasets
            df_combined = pd.concat([
                self.df_ib[['desc']].rename(columns={'desc': 'purpose'}),
                self.df_ik[['tujuan']].rename(columns={'tujuan': 'purpose'})
            ])
            
            # Save to CSV for next stages
            df_combined.to_csv('combined_data.csv', index=False)
            print("Combined data saved to combined_data.csv")

# Main execution
if __name__ == '__main__':
    file_path = 'dataset.xlsx'
    business_phase = BusinessUnderstandingPhase(file_path)
    business_phase.load_and_explore_data()
    business_phase.save_processed_data()
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

class DataPreparationPhase:
    def __init__(self, input_file='combined_data.csv'):
        self.df = pd.read_csv(input_file)
        self.processed_data = None
        self.vectorizer = None
    
    def preprocess_text(self, text):
        """
        Preprocess Indonesian text
        """
        # Convert to lowercase
        text = str(text).lower()
        
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Tokenize
        try:
            tokens = word_tokenize(text)
        except:
            tokens = text.split()
        
        # Remove stopwords
        stop_words = set(stopwords.words('indonesian'))
        tokens = [token for token in tokens if token not in stop_words]
        
        return ' '.join(tokens)
    
    def prepare_data(self):
        """
        Preprocess and prepare data for modeling
        """
        # Preprocess text
        self.df['processed_purpose'] = self.df['purpose'].apply(self.preprocess_text)
        
        # TF-IDF Vectorization
        self.vectorizer = TfidfVectorizer(max_features=100)
        X = self.vectorizer.fit_transform(self.df['processed_purpose']).toarray()
        
        # Visualize feature importance
        feature_names = self.vectorizer.get_feature_names_out()
        feature_importance = X.sum(axis=0)
        
        plt.figure(figsize=(12, 6))
        plt.bar(feature_names, feature_importance)
        plt.title('TF-IDF Feature Importance')
        plt.xlabel('Features')
        plt.ylabel('Importance')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
        
        # Save processed data
        np.save('tfidf_features.npy', X)
        
        # Save feature names
        with open('feature_names.txt', 'w') as f:
            f.write('\n'.join(feature_names))
        
        print("Processed data saved as tfidf_features.npy")
        print("Feature names saved in feature_names.txt")
        
        return X

# Main execution
if __name__ == '__main__':
    prep_phase = DataPreparationPhase()
    X = prep_phase.prepare_data()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

class ClusteringAnalysis:
    def __init__(self, features_file='tfidf_features.npy'):
        self.X = np.load(features_file)
        self.feature_names = None
        
        # Load feature names
        with open('feature_names.txt', 'r') as f:
            self.feature_names = f.read().splitlines()
    
    def determine_optimal_clusters(self, max_clusters=10):
        """
        Find optimal number of clusters using silhouette score
        """
        silhouette_scores = []
        
        for n_clusters in range(2, max_clusters + 1):
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            cluster_labels = kmeans.fit_predict(self.X)
            score = silhouette_score(self.X, cluster_labels)
            silhouette_scores.append(score)
        
        # Plot silhouette scores
        plt.figure(figsize=(10, 6))
        plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
        plt.title('Silhouette Scores for Different Cluster Numbers')
        plt.xlabel('Number of Clusters')
        plt.ylabel('Silhouette Score')
        plt.show()
        
        # Find optimal number of clusters
        optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
        print(f"Optimal number of clusters: {optimal_clusters}")
        
        return optimal_clusters
    
    def perform_clustering(self, n_clusters):
        """
        Perform clustering and analyze results
        """
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        cluster_labels = kmeans.fit_predict(self.X)
        
        # Analyze cluster centroids
        centroids = kmeans.cluster_centers_
        
        # Identify top keywords for each cluster
        cluster_keywords = {}
        for i in range(n_clusters):
            cluster_data = centroids[i]
            top_feature_indices = cluster_data.argsort()[-5:][::-1]
            top_keywords = [self.feature_names[idx] for idx in top_feature_indices]
            cluster_keywords[i] = top_keywords
        
        # Save cluster information
        with open('cluster_analysis.txt', 'w') as f:
            f.write("Cluster Analysis Results\n")
            f.write("======================\n\n")
            for cluster, keywords in cluster_keywords.items():
                f.write(f"Cluster {cluster} Top Keywords: {', '.join(keywords)}\n")
        
        print("Cluster analysis saved to cluster_analysis.txt")
        
        return cluster_labels, cluster_keywords

# Main execution
if __name__ == '__main__':
    clustering = ClusteringAnalysis()
    optimal_clusters = clustering.determine_optimal_clusters()
    cluster_labels, cluster_keywords = clustering.perform_clustering(optimal_clusters)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

class ClassificationModeling:
    def __init__(self, 
                 features_file='tfidf_features.npy', 
                 data_file='combined_data.csv'):
        # Load features
        self.X = np.load(features_file)
        
        # Prepare labels
        self.df = pd.read_csv(data_file)
        
        # Create binary label (you might want to customize this)
        self.df['label'] = (self.df['purpose'].str.contains('laptop|komputer', case=False)).astype(int)
        
        self.y = self.df['label'].values
    
    def train_models(self):
        """
        Train multiple classification models with hyperparameter tuning
        """
        # Split the data
        X_train, X_test, y_train, y_test = train_test_split(
            self.X, self.y, test_size=0.2, random_state=42
        )
        
        # Models and their parameter grids
        models = {
            'Random Forest': {
                'model': RandomForestClassifier(),
                'params': {
                    'n_estimators': [50, 100, 200],
                    'max_depth': [None, 10, 20],
                    'min_samples_split': [2, 5, 10]
                }
            },
            'SVM': {
                'model': SVC(),
                'params': {
                    'C': [0.1, 1, 10],
                    'kernel': ['linear', 'rbf'],
                    'gamma': ['scale', 'auto']
                }
            }
        }
        
        # Results storage
        results = {}
        
        # Hyperparameter tuning and evaluation
        for name, config in models.items():
            grid_search = GridSearchCV(
                estimator=config['model'],
                param_grid=config['params'],
                cv=5,
                scoring='accuracy'
            )
            
            grid_search.fit(X_train, y_train)
            
            # Best model predictions
            best_model = grid_search.best_estimator_
            y_pred = best_model.predict(X_test)
            
            # Evaluation
            results[name] = {
                'best_params': grid_search.best_params_,
                'best_score': grid_search.best_score_,
                'test_accuracy': accuracy_score(y_test, y_pred),
                'classification_report': classification_report(y_test, y_pred)
            }
            
            # Confusion Matrix
            cm = confusion_matrix(y_test, y_pred)
            plt.figure(figsize=(8, 6))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
            plt.title(f'Confusion Matrix - {name}')
            plt.ylabel('Actual')
            plt.xlabel('Predicted')
            plt.tight_layout()
            plt.savefig(f'{name}_confusion_matrix.png')
            plt.close()
        
        # Save results
        with open('model_evaluation_results.txt', 'w') as f:
            
            for name, result in results.items():
                f.write(f"{name} Model Results\n")
                f.write("===================\n")
                f.write(f"Best Parameters: {result['best_params']}\n")
                f.write(f"Best CV Score: {result['best_score']}\n")
                f.write(f"Test Accuracy: {result['test_accuracy']}\n")
                f.write("Classification Report:\n")
                f.write(result['classification_report'])
                f.write("\n\n")
        
        print("Model evaluation results saved to model_evaluation_results.txt")
        
        return results

# Main execution
if __name__ == '__main__':
    classification = ClassificationModeling()
    model_results = classification.train_models()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class BusinessOpportunitiesAnalysis:
    # def __init__(self, 
    #              features_file='tfidf_features.npy', 
    #              data_file='combined_data.csv',
    #              cluster_file='cluster_analysis.txt'):
    #     # Load features and data
    #     self.X = np.load(features_file)
    #     self.df = pd.read_csv(data_file)
        
    #     # Read cluster information
    #     with open(cluster_file, 'r') as f:
    #         cluster_content = f.read()
        
    #     # Parse cluster keywords
    #     self.cluster_keywords = self._parse_cluster_keywords(cluster_content)

    def __init__(self, 
             features_file='tfidf_features.npy', 
             data_file='combined_data.csv',
             cluster_file='cluster_analysis.txt'):
    # Load features and data
        try:
            self.X = np.load(features_file)
        except FileNotFoundError:
            print(f"Error: Features file {features_file} not found!")
            self.X = None

        try:
            self.df = pd.read_csv(data_file)
        except FileNotFoundError:
            print(f"Error: Data file {data_file} not found!")
            self.df = None

        # Read cluster information
        try:
            with open(cluster_file, 'r') as f:
                cluster_content = f.read()
            
            # Parse cluster keywords
            self.cluster_keywords = self._parse_cluster_keywords(cluster_content)
        except FileNotFoundError:
            print(f"Error: Cluster analysis file {cluster_file} not found!")
            self.cluster_keywords = {}
    
    def _parse_cluster_keywords(self, cluster_content):
        """
        Parse cluster keywords from the cluster analysis file
        """
        keywords = {}
        lines = cluster_content.split('\n')
        for line in lines:
            if "Cluster" in line and "Top Keywords:" in line:
                try:
                    # Extract the cluster number
                    cluster_part, keywords_part = line.split("Top Keywords:")
                    cluster = int(cluster_part.split("Cluster")[1].strip())
                    # Extract keywords
                    cluster_keywords = [kw.strip() for kw in keywords_part.split(',')]
                    keywords[cluster] = cluster_keywords
                except (IndexError, ValueError) as e:
                    print(f"Warning: Could not parse line - {line}")
        return keywords

    
    def identify_business_opportunities(self):
        """
        Enhance business opportunity identification with more detailed analysis
        """
        opportunities = []
        
        for cluster, keywords in self.cluster_keywords.items():
            # More sophisticated opportunity generation
            for keyword in keywords:
                opportunity = {
                    'cluster': cluster,
                    'keyword': keyword,
                    'business_concept': f"Toko {keyword}",
                    'target_market': 'Mahasiswa (Students)',
                    'market_potential': self._estimate_market_potential(keyword),
                    'competitive_analysis': self._assess_competition(keyword),
                    'revenue_potential': self._estimate_revenue(keyword)
                }
                opportunities.append(opportunity)
        
        # Sort opportunities by market potential
        opportunities_sorted = sorted(opportunities, 
                                      key=lambda x: x['market_potential'], 
                                      reverse=True)
        
        # Visualize top opportunities
        self._visualize_opportunities(opportunities_sorted)
        
        # Save detailed opportunities report
        self._save_opportunities_report(opportunities_sorted)
        
        return opportunities_sorted

    def _estimate_market_potential(self, keyword):
        """
        Estimate market potential based on frequency in dataset
        """
        keyword_count = (self.df['purpose'].str.contains(keyword, case=False)).sum()
        total_entries = len(self.df)
        return (keyword_count / total_entries) * 100

    def _assess_competition(self, keyword):
        """
        Basic competition assessment
        """
        # This would ideally integrate external market research data
        competition_levels = {
            'laptop': 'High',
            'komputer': 'Medium',
            'elektronik': 'High',
            'buku': 'Low'
        }
        return competition_levels.get(keyword, 'Unknown')

    def _estimate_revenue(self, keyword):
        """
        Rough revenue estimation
        """
        revenue_estimates = {
            'laptop': 15000000,  # IDR 15 million
            'komputer': 3000000,  # IDR 3 million
            'elektronik': 2000000,  # IDR 2 million
            'buku': 15000  # IDR 15 thousand
        }
        return revenue_estimates.get(keyword, 0)

    def _visualize_opportunities(self, opportunities):
        """
        Create visualization of top business opportunities
        """
        plt.figure(figsize=(12, 6))
        top_5_opportunities = opportunities[:5]
        
        plt.bar(
            [opp['business_concept'] for opp in top_5_opportunities],
            [opp['market_potential'] for opp in top_5_opportunities]
        )
        plt.title('Top 5 Business Opportunities')
        plt.xlabel('Business Concept')
        plt.ylabel('Market Potential (%)')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.savefig('top_opportunities.png')
        plt.close()

    def _save_opportunities_report(self, opportunities):
        """
        Generate a comprehensive opportunities report
        """
        with open('business_opportunities_report.md', 'w') as f:
            f.write("# Business Opportunities Analysis Report\n\n")
            f.write("## Top Business Opportunities\n\n")
            
            for i, opp in enumerate(opportunities[:10], 1):
                f.write(f"### {i}. {opp['business_concept']}\n")
                f.write(f"- **Cluster**: {opp['cluster']}\n")
                f.write(f"- **Target Market**: {opp['target_market']}\n")
                f.write(f"- **Market Potential**: {opp['market_potential']:.2f}%\n")
                f.write(f"- **Competitive Landscape**: {opp['competitive_analysis']}\n")
                f.write(f"- **Estimated Annual Revenue**: IDR {opp['revenue_potential']:,}\n\n")

# Main execution
def main():
    try:
        # Initialize the Business Opportunities Analysis
        business_opportunities = BusinessOpportunitiesAnalysis()
        
        # Identify and analyze business opportunities
        opportunities = business_opportunities.identify_business_opportunities()
        
        # Print top 5 opportunities to console
        print("Top 5 Business Opportunities:")
        for i, opp in enumerate(opportunities[:5], 1):
            print(f"{i}. {opp['business_concept']}")
            print(f"   Market Potential: {opp['market_potential']:.2f}%")
            print(f"   Estimated Revenue: IDR {opp['revenue_potential']:,}")
            print(f"   Competitive Landscape: {opp['competitive_analysis']}\n")
        
        print("Full report saved to 'business_opportunities_report.md'")
        print("Visualization saved to 'top_opportunities.png'")
    
    except Exception as e:
        print(f"An error occurred: {e}")
        import traceback
        traceback.print_exc()

# Ensure the script can be run directly or imported
if __name__ == '__main__':
    main()
```
